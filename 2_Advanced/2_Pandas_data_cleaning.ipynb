```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Datacleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# fill Nan values with medians 
salary_year_median = df['salary_year_avg'].median()
salary_year_median
```

```python
salary_hour_median = round(df['salary_hour_avg'].median(), 2)
salary_hour_median
```

```python
df_filled = df

df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(salary_year_median)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(salary_hour_median)
```

```python
df_filled['salary_year_avg']
```

```python
df_filled['salary_hour_avg']
```

```python
df_unique = df_filled

df_unique = df_unique.drop_duplicates()

print(f'Total number of rows      :{len(df_filled)}')
print(f'Number of rows remaining  :{len(df_unique)}')
print(f'Number of rows dropped    :{len(df_filled)-len(df_unique)}')
```

```python
df_unique = df_filled

df_unique = df_unique.drop_duplicates(subset=['job_title','company_name'])

print(f'Total number of rows      :{len(df_filled)}')
print(f'Number of rows remaining  :{len(df_unique)}')
print(f'Number of rows dropped    :{len(df_filled)-len(df_unique)}')
```
