# Copy()

```python
dataset = load_dataset('lukebarousse/data_jobs')
df_orginal = dataset['train'].to_pandas()
```

```python
# Until now instead of using this method we were simply assigning dataframes
df_altered = df_orginal
# below we will see the problem of doing this way
```

```python
# just like before we will fill the Nan values with median
median_salary = df_altered['salary_year_avg'].median()
median_salary
```

```python
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary)
df_altered['salary_year_avg']
```

```python
# now when we call the orginal dataframe we think that it might still have Nan values but it doesnt it also changed
df_orginal['salary_year_avg']
```

```python
id(df_altered) == id(df_orginal) # see both of the dataframes have the same id
```

```python
# so inorder to keep the orginal dataframe intact we must use copy()
dataset = load_dataset('lukebarousse/data_jobs')
df_orginal = dataset['train'].to_pandas()

df_altered = df_orginal.copy()
```

```python
# now we do the same thing
median_salary = df_altered['salary_year_avg'].median()
median_salary
```

```python
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary)
df_altered['salary_year_avg']
```

```python
df_orginal['salary_year_avg']
```

```python
print(f'id number of df_orginal     : {id(df_orginal)}')
print(f'id number of df_altered     : {id(df_altered)}')
print(f'Are both the dataframe same?: {id(df_orginal) == id(df_altered)}')
```
