## Exploratory Data Analysis ##

```python
# importing libraries
import pandas as pd
import ast
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

### Filter US Data analyst Roles ###

```python
df_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')]
df_US
```

```python
df_plot = df_US['job_location'].value_counts().head(10).to_frame()
sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.xlabel('Number of Job Postings')
plt.ylabel('')
plt.title('Counts of Job locations for Data Analyst Roles')
plt.show()
```

```python
dict_columns = {
    'job_work_from_home': 'Work From Home',
    'job_no_degree_mention': 'Degree required Jobs',
    'job_health_insurance': 'Health insurance required Jobs'
}
fig, ax = plt.subplots(1,3)
fig.set_size_inches((12,5))

for i,(column,title) in enumerate(dict_columns.items()):
    ax[i].pie(df_US[column].value_counts(), labels=['False','True'], autopct= '%1.1f%%', startangle=90)
    ax[i].set_title(title)
plt.tight_layout()
plt.show()
```

```python
df_plot = df_US['company_name'].value_counts().head(10).to_frame()
```

```python
sns.set_theme(style='ticks')
sns.barplot(data= df_plot, x='count', y='company_name', hue='count', palette='light:b', legend=False)
sns.despine()
plt.xlabel('Number of Job Postings')
plt.ylabel('')
plt.title('Counts of US Companies for Data Analyst Roles')
plt.show()
```
