````python
# symmetric difference ^  returns all items which are not repeated
day1_users = {"A", "B", "C"}
day2_users = {"C", "D", "E"}

one_time_users = day1_users ^ day2_users
one_time_users
````

````python
class lukelist:
  def __init__(self):
      """ Initialize an empty list"""
      self._items = []

  def add(self, item):
    """Add an item to the end of the list."""
    self._items.append(item)

  def __getitem__(self, index):
    """Retrieve an item by index"""
    return self._items[index]

  def __setitem__(self, index, value):
    """Set an item at a specific index"""
    self._items[index] = value

  def __repr__(self):
    """Return a string representation of the List"""
    return str(self._items)

  def __len__(self):
    """Return the length of the list"""
    return len(self._items)
````

````python
def calculate_salary(base_salary, bonus_rate =.1):
  """
  calculate the total salary based on the base salary and bonus rate.

  Args:
    base_salary (float): The base salary.
    bonus_rate (float): The bonus rate. Default is .1

  Returns:
    float: The total Salary
  """
  return base_salary * (1 + bonus_rate)

def calculate_bonus(total_salary, base_salary):
  """
  Calculate the bonus based on the total salary and base salary.

  Args:
    total_salary (float): The total salary.
    base_salary (float): The base salary.

  Returns:
    float: The bonus amount.
  """
  return (total_salary - base_salary) / base_salary
````

````python
# we can make this into a class
class BaseSalary:
  def __init__(self, base_salary, bonus_rate=.1, symbol="$"):
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.symbol = symbol
    self.total_salary = base_salary * (1 + bonus_rate)
    self.bonus = self.total_salary - self.base_salary

  def __repr__(self):
    return f'{self.symbol}{self.base_salary:,.0f}'

  def show_salary(self):
    return f'{self.symbol}{self.total_salary:,.0f}'

  def show_bonus(self):
    return f'{self.symbol}{self.bonus:,.0f}'
````

````python
class joblist:
  def __init__(self, jobs):
    self.jobs = jobs

  def __str__(self):
    jobs_str = "Jobs: \n"
    for job in self.jobs:
      jobs_str += f'- {job["job_title"]} : {job["job_skills"]}'
    return jobs_str

  def split_skills(self):
    for job in self.jobs:
      if isinstance(job['job_skills'], str):
        job['job_skills'] = job['job_skills'].split(",")
    return job['job_skills']

  def display_jobs(self):
    for job in self.jobs:
      print(f'Job Title: {job['job_title']}')
      print("Required skills: ")
      for skill in job["job_skills"]:
        print(f' - {skill}')

  def filter_by_skill(self, skill):
    filtered_jobs = [job['job_title'] for job in self.jobs if skill in job['job_skills']]
    print(filtered_jobs)

  def skills_are_split(self):
    return all(isinstance(job['job_skills'], list) for job in self.jobs)
````

````python
job_list = joblist(jobs)
print(job_list)

job_list.split_skills()
print(job_list)

job_list.display_jobs()

job_list.filter_by_skill("Python")

job_list.skills_are_split()
````

````python
class JobList:
  def __init__(self,jobs):
    self.jobs = jobs

  def __str__(self):
    job_str = "Jobs:"
    for job in self.jobs:
      job_str += f'\n - {job["job_title"]}: {job["job_skills"]}'
    return job_str

  def split_skills(self):
    for job in self.jobs:
      if isinstance(job['job_skills'], str):
         job['job_skills'] = job['job_skills'].split(", ")

  def display_jobs(self):
    for job in self.jobs:
      print(f'Job Title: {job["job_title"]}')
      print("Required skills: ")
      for skill in job['job_skills']:
         print(f' - {skill}')

  def filter_by_skill(self,skill):
    return [job['job_title'] for job in self.jobs if skill in  job['job_skills']]

````
